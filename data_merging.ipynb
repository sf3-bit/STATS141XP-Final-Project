```python
import pandas as pd
```

Read in Pitcher Stats data

```python
# read in pitcher stats data
pitcher_stats_df = pd.read_csv("compiled_pitcherGame.csv")
pitcher_stats_df.head()
```

Read in WAR stats data

```python
# read in all WAR dataframes
war_df_list = [pd.read_csv(f"{year}.csv") for year in range(2010, 2025)]
war_df_list[0].head()
```

```python
# for each WAR dataframe, drop every column except "Name" and "Pit WAR"
war_df_list = [war_df[["Name", "Pit WAR"]] for war_df in war_df_list]
# for each WAR dataframe, rename "Pit WAR" column to "WAR" column
for war_df in war_df_list:
    war_df.rename(columns = {"Pit WAR" : "WAR"}, inplace = True)
war_df_list[0].head()
```

```python
# for each WAR dataframe, remove any rows with NA, since if there is no WAR data, there is no point in joining
for war_df in war_df_list:
    war_df.dropna(inplace = True)
war_df_list[0].head()
```

We want to conduct the join on the `Name` column, so let's view how the elements look in the pitcher stats dataframe

```python
pitcher_stats_df["name"].head()
```

The `name` column of the pitcher stats dataframe is in "last name, first name" format, we can change it to "first name, last name" format...

this is done to match the format of the combined WAR dataframe

```python
# change to "first name, last name" format
pitcher_stats_df["name"] = pitcher_stats_df['name'].str.split(', ').str[::-1].str.join(' ')
pitcher_stats_df["name"].head()
```

```python
# for each year, join the pitcher stats of that particular year, with it's corresponding WAR, if it exists
# then, concatenate the merged dataframe of each year
df_merge = pd.DataFrame()

for i in range(len(war_df_list)):
    year = i + 2010
    year_pitcher_stats_df = pitcher_stats_df[pitcher_stats_df["agg_Date"].astype(str).str[:4] == str(year)]
    year_war_df = war_df_list[i]
    year_merged_df = pd.merge(left = year_pitcher_stats_df, right = year_war_df, left_on = "name", right_on = "Name", how = "inner")
    df_merge = pd.concat([df_merge, year_merged_df], axis = 0)
df_merge.head()
```

```python
# write final dataset to csv file
df_merge.to_csv("merged_dataset.csv", index = False)
```
