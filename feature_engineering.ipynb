```python
import pandas as pd
```

```python
# read in merged data
df = pd.read_csv("merged_data.csv")
df.head()
```

Feature Engineering - Feature Selection

```python
# construct non-game stat dataframe

# select features
non_game_stats_features = ["name", "ageYrs", "seniorityYrs", "RHP"]

# create non-game stat dataframe
non_game_stats_df = df[non_game_stats_features]

non_game_stats_df.head()
```

```python
# construct game stat dataframe

# select features
game_stats_features = ["K", "W", "HR", "outs", "x", "agg_Date"]

# create game stat dataframe
game_stats_df = df[game_stats_features]

game_stats_df.head()
```

```python
# construct war dataframe
war_stats_df = pd.DataFrame(df["WAR"])
war_stats_df.head()
```

```python
# construct final dataset

# concatenate non-game stat dataframe, game stat dataframe, and WAR dataframe
feature_df = pd.concat([non_game_stats_df, game_stats_df, war_stats_df], axis = 1)

feature_df.head()
```

Feature Engineering - Feature Transformations
- Transform data into such a form that we expect to be input for the model

```python
# change date column to year column
feature_df["agg_Date"] = feature_df["agg_Date"].astype(str).str[:4].astype(int)
feature_df.rename(columns = {"agg_Date" : "year"}, inplace = True)
feature_df.head()
```

```python
# divide game stats (number of notable pitching events) by the overall total number of pitching events
feature_df[["K", "W", "HR", "outs"]] = feature_df[["K", "W", "HR", "outs"]].div(feature_df["x"], axis = 0)
feature_df.drop(columns=["x"], inplace = True)
feature_df.head()
```

```python
# group by (player and year), then compute aggregate stats, which will act as input to the model

# custom aggregate function to compute mean, mode, etc.
def custom_agg(group):
    return pd.Series({
        "avg_ageYrs" : group["ageYrs"].mean(),
        "avg_seniorityYrs" : group["seniorityYrs"].mean(),
        "RHP" : group["RHP"].mode()[0], # mode returns a series
        "avg_K" : group["K"].mean(),
        "avg_W" : group["W"].mean(),
        "avg_HR" : group["HR"].mean(),
        "avg_outs" : group["outs"].mean(),
        "WAR" : group["WAR"].mode()[0]
    })

final_df = feature_df.groupby(["name", "year"]).apply(custom_agg).reset_index()
final_df.head()
```

```python
final_df.to_csv("final_dataset.csv", index = False)
```
